```python

import pandas as pd
from sklearn.datasets import load_iris
```

# 1. Collect Dataset - Load Iris Flower Dataset from sklearn library

iris = load_iris()

```python
iris.feature_names # Print Iris Flower feature names
```

```python
iris.target_names
```

# 2. Prepare the dataset

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
# In target column 0, 1, 2 are the encoded values for setosa, versicolor and virginica respectively.
```

```python
df[df.target == 0].head()
```

```python
df[df.target == 1].head()
```

```python
df[df.target == 2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head() # Add a new column with flower names instead of encoded values in target column.
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# 3. Split the dataset into Training and Testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis = 'columns') # Drop prediction target and flower name columns to get feature data.
y = df.target # Prediction target is stored in target column.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

# 4. Train the Model

from sklearn.svm import SVC

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

# 5. Predict the Accuracy by Testing Dataset

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8, 3.0, 1.5, 0.3]])
```

# Tune Parameters

## Regularization (C)

model_C = SVC(C = 1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)

```python
model_C = SVC(C = 10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

## Gama

model_g = SVC(gamma = 10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)

## Kernel

model_linear_kernal = SVC(kernel = 'linear')
model_linear_kernal.fit(X_train, y_train)

```python
model_linear_kernal.score(X_test, y_test)
```
