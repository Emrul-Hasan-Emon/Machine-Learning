# Predicting If a Person Will Buy Life Insurnace Based on Age

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

# 1. Collect Dataset

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, marker='+', color='red')
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.title('Insurance Data')
plt.show()
```

# 2. Prepare DataSET 
## Splitting the Total Dataset into 80/20 - Training Data Set - 80%, Testing Data Set - 20%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, test_size=0.2)
```

```python
X_test
```

## Importing Logistic Regression Model   

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

# 3. Train The Model

```python
model.fit(X_train, y_train)
```

# 4. Predict The Model Based on Testing DataSet

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
model.predict_proba(X_test)
```

## Get the Accuracy

```python
model.score(X_test,y_test)
```

## model.coef_ indicates value of m in y = m * x + b Equation

```python
coefficient = model.coef_[0][0]
coefficient
```

## model.intercept_ indicates value of b in y = m * x + b equation

```python
intercept = model.intercept_[0]
intercept
```

# Sigmoid Function
## Predict value by Sigmoid Function (Math) manually

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(age):
    z = coefficient * age + intercept
    y = sigmoid(z)
    return y
```

```python
age = 35
prediction_function(age)
```
