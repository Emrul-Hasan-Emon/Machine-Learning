```python
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
```

# 1. Collect Dataset - Digits dataset from sklearn

```python
digits = load_digits()
```

```python
plt.gray() 
for i in range(5):
    plt.matshow(digits.images[i]) 
```

```python
dir(digits)
```

# 2. Split the dataset into training, testing dataset. Training Dataset - 80%, Testing Dataset - 20%.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2)
```

# 3. Train the model with Training Dataset

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

## Finding Accuracy

```python
model.score(X_test, y_test)
```

```python
model.predict(digits.data[0:5])
```

# 4. Predict the Model with Testing Dataset

```python
y_predicted = model.predict(X_test)
y_predicted
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```
